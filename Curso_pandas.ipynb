## Series

Las *series* de Pandas son un tipo de objeto semejante a un *vector* con 칤ndices, donde cada 칤ndice es un identificador (que puede ser un literal o un n칰mero) que permite acceder a determinada posici칩n de la *serie*. Para crear una serie podemos utilizar el comando *pd.Series()* donde tendremos distintas combinaciones de argumentos.

```python
dicc = {"primero" : 1 , "segundo" : 2, "tercero" : 3}
serie_desde_diccionario = pd.Series(dicc)
print(serie_desde_diccionario)
```

```python
lista = [0,1,2]
serie_desde_lista = pd.Series(lista)
print(serie_desde_lista)
```

```python
arreglo = np.array([0,1, 2])
serie_desde_arreglo = pd.Series(arreglo)
print(serie_desde_arreglo)
```

Notamos que en nuestro primer caso los *칤ndices* de nuestra serie son literales, mientras que en los otros dos, son n칰meros enteros. Esto sucede por que el crear una serie a trav칠s de un diccionario asigna inmediatamente el 칤ndice a un valor con su correspondiente llave. Si queremos cambiar el 칤ndice de una serie, podemos utilizar los siguientes comandos.

```python
print(serie_desde_arreglo.index) #Obtenci칩n de los 칤ndices de la serie
serie_desde_arreglo.index = ["primero","segundo","tercero"] #reemplazo de los 칤ndices
print(serie_desde_arreglo) #Serie con 칤ndices cambiados
```

## Agrupando datos 游빌

[Inicio 郊쑇(#Indice)

Un aspecto interesante de trabajar con tablas, es poder agrupar los datos de acuerdo con alguna columna. Esto permitir치 caracterizar de mejor manera nuestros datos.  
Para poder trabajar con los datos agregados ser치 importante identificar sobre que tipo de dato estamos agregando, por ejemplo, en nuestro caso podr칤amos agrupar datos a trav칠s de la columna "EDUCATION", lo cual formar치 otra tabla que tendr치 como 칤ndices los tipos de Educaci칩n y como columnas las mismas de la tabla original pero, se calcularan nuevos campos dependiendo de como queramos agregar los datos que aparecen en cada repetici칩n de los tipos de educaci칩n. Para ejemplificarlo mejor veamos como se hace la agrupaci칩n de una tabla en pandas y comparemosla a m칠tricas extraidas directamente de la misma tabla.

```python
# Para agrupar los datos, primero se crea un objeto de agrupaci칩n, indicando cual es la columna que se utilizar치 para agrupar.
datos_grouped = tabla.groupby("EDUCATION")
# Luego podemos utilizar distintos tipos de funciones de agregaci칩n, lo que har치n estas funciones ser치 tomar todas las columnas en que determinado valor de la columna "EDUCATION" se repita
# y agregar los datos de las otras columnas dependiendo de la funci칩n especificada. Ejemplo de estas funciones son la suma de datos, el calculo de la media, el calculo de la mediana, etc... En este caso, 
# agregaremos el resto de las columnas con la funci칩n suma especificada por el atributo ".sum"
datos_grouped = datos_grouped.sum()
datos_grouped
```

```python
# Como podemos ver hemos agrupado nuestros datos a trav칠s del nivel de educaci칩n, donde por ejemplo en la columna "BILL_AMT4" podemos ver a deuda total de cada grupo de personas
# con distintos tipos de educaci칩n. Sin embargo, hay columnas a las cuales no hace sentido calcular la suma total, por ejemplo, la edad. En este caso, tenemos otro tipo de atributo para 
# nuestra agrupaci칩n, lo que haremos ser치 trabajar con la funci칩n *mean*. Podemos hacer aquello muy similar a como lo hicimos anteriormente.
datos_grouped = tabla.groupby("EDUCATION").mean()
datos_grouped
#Podemos observar entonces que el promedio de edad entre las personas de cada grupo tiene una media con rango de 30 a 45 a침os.
```

Tambi칠n es posible trabajar con funciones de agregaciones diferentes *dependiendo de la columna que estes trabajando*. Para esto se ocupa la funci칩n ".agg()" donde el argumento de la funci칩n viene a ser _**un diccionario**_ donde las claves del diccionario son los nombres de las columnas a agregar y los valores vienen a ser las funciones que utilizar치s.

```python
dict_funciones = {"SEX" : pd.Series.mode,
                  "AGE" : "mean",
                  "PAY_AMT2" : sum}
# En este caso, la columna "SEX" se agrupar치 por moda, la columna "AGE" se agrupar치 por la media
# y la columna "PAY_AMT2" se agrupara sumando los datos.
datos_grouped = tabla.groupby("EDUCATION").agg(dict_funciones)
datos_grouped
```

Podemos extraer entonces de los datos de nuestra tabla que contrario al resto de niveles educacionales, hay mas personas con "SEX" == 1 en el nivel educacional 0.
